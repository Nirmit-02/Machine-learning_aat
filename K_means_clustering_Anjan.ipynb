K MEANS CLUSTERING

K-means clustering is a popular unsupervised machine learning algorithm used to partition a dataset into k distinct, non-overlapping subsets (clusters). Each data point belongs to the cluster with the nearest mean, serving as a prototype of the cluster.
The following dataset that was chosen is a cancer dataset and the attributes are chosen based on the characteristic of the tumour
the attributes chosen are "smoothness" and "concativity"
Smoothness : It typically measures the regularity of the tumor perimeter. A lower smoothness value indicates a more regular and smooth boundary, while a higher value indicates a more irregular and jagged boundary.
Concavity: This refers to the extent to which the tumor surface has concave parts (indented or inward-curving parts). Higher concavity values indicate more pronounced concavities in the tumor boundary, which can be a characteristic of malignancy.

```python
import pandas as pd
from sklearn.cluster import KMeans as K
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Cancer_Data_Anjan.csv')
```

```python
df
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans as K
from matplotlib import pyplot as plt



# Plot the scatter plot
plt.scatter(df['smoothness_mean'], df['concavity_mean'])
plt.xlabel('Smoothness')
plt.ylabel('Concavity')
plt.title('Scatter Plot of Smoothness vs Concavity')
plt.show()
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
  km = K(n_clusters = k)
  km.fit(df[['smoothness_mean', 'concavity_mean']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse)
```

```python
km = K(n_clusters=4)
km
```

```python
y=km.fit_predict(df[['smoothness_mean', 'concavity_mean']])
y
```

```python
df['cluster']=y
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
plt.scatter(df1.smoothness_mean,df1['concavity_mean'], color = 'green')
plt.scatter(df2.smoothness_mean,df2['concavity_mean'], color = 'blue')
plt.scatter(df3.smoothness_mean,df3['concavity_mean'], color = 'red')
plt.scatter(df4.smoothness_mean,df4['concavity_mean'], color = 'pink')
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
plt.scatter(df1.smoothness_mean,df1['concavity_mean'], color = 'green')
plt.scatter(df2.smoothness_mean,df2['concavity_mean'], color = 'blue')
plt.scatter(df3.smoothness_mean,df3['concavity_mean'], color = 'red')
plt.scatter(df4.smoothness_mean,df4['concavity_mean'], color = 'pink')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='k',marker='*',label='centroid')
```
