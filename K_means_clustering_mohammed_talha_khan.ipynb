K-means clustering is a popular unsupervised machine learning algorithm used to partition a dataset into k distinct, non-overlapping subsets (clusters). Each data point belongs to the cluster with the nearest mean, serving as a prototype of the cluster.
The following dataset that was chosen is a cancer dataset and the attributes are chosen based on the characteristic of the tumour
the attributes chosen are "perimeter" and "area"
Perimeter : The perimeter of a tumor refers to the length of its boundary. In a two-dimensional image, this is the total length of the outline that encircles the tumor
Area : The area of a tumor is the number of pixels or physical units (e.g., square millimeters) that the tumor occupies in a two-dimensional image.

```python
import pandas as pd
from sklearn.cluster import KMeans as K
from matplotlib import pyplot as plt
```

```python
df = pd. read_csv('Cancer_Data_Talha.csv')
```

```python
df
df.head()
```

```python
plt.scatter(df['perimeter_se'], df['area_se'])
plt.xlabel('perimeter_se')
plt.ylabel('area_se')
plt.title('Scatter Plot of perimeter_se vs area_se')
plt.show()
```

```python
k_range = range(1,8)
sse = []
for k in k_range:
  km = K(n_clusters = k)
  km.fit(df[['perimeter_se', 'area_se']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse)
```

```python
km= K(n_clusters=3)
km
```

```python
y=km.fit_predict(df[['perimeter_se', 'area_se']])
y
```

```python
df['cluster']=y
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.perimeter_se,df1['area_se'], color = 'red')
plt.scatter(df2.perimeter_se,df2['area_se'], color = 'blue')
plt.scatter(df3.perimeter_se,df3['area_se'], color = 'green')
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.perimeter_se,df1['area_se'], color = 'green')
plt.scatter(df2.perimeter_se,df2['area_se'], color = 'blue')
plt.scatter(df3.perimeter_se,df3['area_se'], color = 'red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='k',marker='*',label='centroid')
```
