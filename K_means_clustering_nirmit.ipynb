K-means clustering is a popular unsupervised machine learning algorithm used to partition a dataset into k distinct, non-overlapping subsets (clusters). Each data point belongs to the cluster with the nearest mean, serving as a prototype of the cluster.
The following dataset that was chosen is a cancer dataset and the attributes are chosen based on the characteristic of the tumour
the attributes chosen are "radius" and "texture"
Radius: The radius of a tumor is typically defined as the distance from the center of the tumor to its boundary. In some datasets, the radius might refer to the mean radius, which is the average of the distances from the center to the boundary across all points on the perimeter.
Texture: Texture measures the variation in the intensity of pixels within the tumor. It captures the smoothness or roughness of the tumor surface and can include various statistical metrics, such as standard deviation of pixel intensity, contrast, and entropy.

```python
import pandas as pd
from sklearn.cluster import KMeans as K
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Cancer_Data_nirmit.csv')
```

```python
df
df.head()
```

```python
plt.scatter(df['radius_mean'], df['texture_mean'])
plt.xlabel('Radius')
plt.ylabel('Texture')
plt.title('Scatter Plot of Radius vs Texture')
plt.show()
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
  km = K(n_clusters = k)
  km.fit(df[['radius_mean', 'texture_mean']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse)
```

```python
km = K(n_clusters=4)
km
```

```python
y=km.fit_predict(df[['radius_mean', 'texture_mean']])
y
```

```python
df['cluster']=y
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
plt.scatter(df1.radius_mean,df1['texture_mean'], color = 'green')
plt.scatter(df2.radius_mean,df2['texture_mean'], color = 'blue')
plt.scatter(df3.radius_mean,df3['texture_mean'], color = 'red')
plt.scatter(df4.radius_mean,df4['texture_mean'], color = 'pink')
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
plt.scatter(df1.radius_mean,df1['texture_mean'], color = 'green')
plt.scatter(df2.radius_mean,df2['texture_mean'], color = 'blue')
plt.scatter(df3.radius_mean,df3['texture_mean'], color = 'red')
plt.scatter(df4.radius_mean,df4['texture_mean'], color = 'pink')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='k',marker='*',label='centroid')
```
